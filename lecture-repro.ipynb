```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b = 3, 2

X = np.random.rand(100)
Y = a * X + b + (np.random.rand(100) - np.random.rand(100)) / 2
```

```python
def model(a, b, x):
    return a * x + b
```

```python
def loss(a, b, X, Y):
    loss = 0
    for x, y in zip(X, Y):
        y_pred = a * x + b
        loss += (y_pred - y) ** 2
    return loss

def grad_loss(model, X, Y):
    dLda, dLdb = 0, 0
    for x, y in zip(X, Y):
        y_pred = model(x)
        dLda += 2 * (y_pred - y) * x
        dLdb += 2 * (y_pred - y)
    return (dLda, dLdb)
```

```python
a0, b0 = 8, -4
X_uniform = np.arange(0, 1, 0.05)
```

```python
plt.scatter(X, Y)
plt.plot(X_uniform, [model(a0, b0, x) for x in X_uniform], color='orange')
plt.plot()
```

```python
learning_rate = 0.0005
iterations = 10000
a_pred = a0
b_pred = b0

for i in range(iterations):
    m = lambda x: model(a_pred, b_pred, x)
    print(f'loss = {loss(a_pred, b_pred, X, Y)}')
    (dLda, dLdb) = grad_loss(m, X, Y)
    a_pred -= learning_rate * dLda
    b_pred -= learning_rate * dLdb

print(f'true (a, b) = ({a}, {b})')
print(f'calc (a, b) = ({a_pred}, {b_pred})')
```

```python
plt.scatter(X, Y)
plt.plot(X_uniform, [model(a_pred, b_pred, x) for x in X_uniform], color='orange')
plt.plot()
```
